```python
df = pd.read_excel("dataset/Pokemon.xlsx")
df.head()
```

```python
# Análisis descriptivo del dataset
print("Dimensiones del dataset:", df.shape)
print("\nColumnas y tipos de datos:")
print(df.dtypes)
print("\nValores faltantes por columna:")
print(df.isna().sum())
print("\nEstadísticas numéricas generales:")
display(df.describe(include='all').T)
print("\nConteo de valores únicos para columnas categóricas:")
for col in df.select_dtypes(include=['object', 'category']).columns:
    print(f"- {col}: {df[col].nunique()} únicos")

print("\nResumen de correlación entre variables numéricas:")
num_df = df.select_dtypes(include=["number"])
if num_df.shape[1] > 0:
    corr = num_df.corr()
    display(corr)
else:
    print("No hay columnas numéricas para calcular la correlación.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Crear carpeta para imágenes
os.makedirs('images', exist_ok=True)

sns.set(style='whitegrid', palette='husl')

num_cols = df.select_dtypes(include=['number']).columns.tolist()
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# 1. Matriz de correlación de estadísticas (stats relevantes)
stats_cols = [col for col in num_cols if col.lower() in ['hp', 'attack', 'defense', 'sp. atk', 'sp. def', 'speed']]
if len(stats_cols) >= 2:
    plt.figure(figsize=(10, 8))
    corr_matrix = df[stats_cols].corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, 
                square=True, linewidths=1, cbar_kws={"shrink": 0.8}, fmt='.2f')
    plt.title('Correlación entre Estadísticas de Pokémon')
    plt.tight_layout()
    plt.savefig('images/01_correlacion_stats.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Identificar tradeoffs (correlaciones negativas)
    print("\n=== ASOCIACIONES CLAVE ===")
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            corr_val = corr_matrix.iloc[i, j]
            col1, col2 = corr_matrix.columns[i], corr_matrix.columns[j]
            if abs(corr_val) > 0.5:
                print(f"{col1} <-> {col2}: {corr_val:.3f}")

# 2. Distribución de stats por tipo
if 'Type 1' in cat_cols and len(stats_cols) > 0:
    primary_type = 'Type 1'
    fig, axes = plt.subplots(2, 3, figsize=(16, 10))
    axes = axes.flatten()
    
    for idx, stat in enumerate(stats_cols[:6]):
        sns.boxplot(data=df, x=primary_type, y=stat, ax=axes[idx], palette='Set2')
        axes[idx].set_title(f'{stat} por Tipo')
        axes[idx].set_xticklabels(axes[idx].get_xticklabels(), rotation=45, ha='right')
    
    plt.tight_layout()
    plt.savefig('images/02_stats_por_tipo.png', dpi=300, bbox_inches='tight')
    plt.show()

# 3. Relación Ataque vs Defensa (análisis de roles)
if 'Attack' in stats_cols and 'Defense' in stats_cols:
    plt.figure(figsize=(12, 7))
    
    if 'Type 1' in cat_cols:
        types = df['Type 1'].unique()[:5]  # Primeros 5 tipos para claridad
        for ptype in types:
            mask = df['Type 1'] == ptype
            plt.scatter(df[mask]['Attack'], df[mask]['Defense'], 
                       label=ptype, alpha=0.6, s=80)
    else:
        plt.scatter(df['Attack'], df['Defense'], alpha=0.6, s=80)
    
    plt.xlabel('Attack')
    plt.ylabel('Defense')
    plt.title('Posicionamiento Ofensivo vs Defensivo de Pokémon')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('images/03_ataque_vs_defensa.png', dpi=300, bbox_inches='tight')
    plt.show()

# 4. Stats totales vs velocidad (hipótesis: tradeoff con velocidad)
if 'Speed' in stats_cols:
    total_stats = df[stats_cols].sum(axis=1)
    plt.figure(figsize=(10, 6))
    plt.scatter(total_stats, df['Speed'], alpha=0.6, s=100, c=df['Speed'], cmap='viridis')
    z = np.polyfit(total_stats, df['Speed'], 1)
    p = np.poly1d(z)
    plt.plot(total_stats, p(total_stats), "r--", linewidth=2, label='Tendencia')
    plt.xlabel('Stats Totales')
    plt.ylabel('Velocidad')
    plt.title('¿Hay tradeoff entre Stats Totales y Velocidad?')
    plt.colorbar(label='Speed')
    plt.legend()
    plt.tight_layout()
    plt.savefig('images/04_stats_totales_vs_velocidad.png', dpi=300, bbox_inches='tight')
    plt.show()

print("\n✓ Imágenes guardadas en la carpeta 'images'")
```
